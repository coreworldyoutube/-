```python
import shutil
import os

# 指定されたディレクトリを再帰的に削除する関数
def delete_folder(path):
    if os.path.exists(path):
        shutil.rmtree(path)  # rmtreeでディレクトリを削除
        print(f"{path} フォルダが削除されました。")
    else:
        print(f"{path} は存在しません。")

# メイン処理
def main():
    folder_to_delete = "A"  # 削除したいフォルダ名
    delete_folder(folder_to_delete)  # フォルダ削除処理

# 実行
main()
```

```python
import os
import hjson

# MOD情報を入力する関数
def get_mod_info():
    mod_info = {}

    mod_info["name"] = input("MODの名前を入力してください: ")
    mod_info["displayName"] = input("MODの表示名を入力してください: ")
    mod_info["author"] = input("MODの作者名を入力してください: ")
    mod_info["description"] = input("MODの説明を入力してください: ")
    mod_info["version"] = input("MODのバージョンを入力してください: ")
    mod_info["minGameVersion"] = input("最小ゲームバージョンを入力してください: ")

    dependencies = input("依存関係があればリストで入力してください（例：['mod1', 'mod2']）: ")
    if dependencies:
        mod_info["dependencies"] = hjson.loads(dependencies)

    mod_info["hidden"] = input("MODを隠す場合はTrue、表示する場合はFalseを入力してください: ")

    return mod_info

# アイテム情報を入力する関数
def get_item_info():
    item_info = {}

    item_info["name"] = input("アイテムの名前を入力してください: ")
    item_info["description"] = input("アイテムの説明を入力してください: ")
    item_info["cost"] = input("アイテムのコストを入力してください（数値）: ")
    item_info["color"] = input("アイテムの色（Hexコード、例：#c1d3f5）を入力してください: ")
    item_info["flammability"] = input("アイテムの引火性を入力してください（数値）: ")
    item_info["explosiveness"] = input("アイテムの爆発性を入力してください（数値）: ")
    item_info["radioactivity"] = input("アイテムの放射能を入力してください（数値）: ")
    item_info["charge"] = input("アイテムの電荷を入力してください（数値）: ")

    return item_info

# ブロック情報を入力する関数
def get_block_info():
    block_info = {}

    # 最初にtype（コンベアなど）を選択
    block_info["type"] = input("ブロックのタイプを入力してください（例：Conveyor、Turretなど）: ")

    block_info["name"] = input("ブロックの名前を入力してください: ")
    block_info["description"] = input("ブロックの説明を入力してください: ")
    block_info["health"] = int(input("ブロックの耐久値を入力してください: "))

    block_info["requirements"] = []
    while True:
        req = input("requirements（例：copper/2）の形式で入力してください。終了したい場合は、'End' を入力してください: ")
        if req.lower() == 'end':
            break
        else:
            block_info["requirements"].append(req)

    block_info["category"] = input("カテゴリー（例：distribution）を入力してください: ")

    # コンベアの場合、速度に関する情報を求める
    if block_info["type"].lower() == "conveyor":
        block_info["speed"] = float(input("コンベアの速度を入力してください（小数値）: "))
        block_info["displayedSpeed"] = float(input("表示されるコンベアの速度を入力してください（小数値）: "))

    # 研究名の入力
    block_info["research"] = input("研究名を入力してください: ")

    return block_info

# HJSON形式で情報を書き込む関数
def write_to_hjson(project_name, mod_info, item_info, block_info):
    # プロジェクトのディレクトリ作成
    if not os.path.exists(project_name):
        os.mkdir(project_name)
    if not os.path.exists(f"{project_name}/content"):
        os.mkdir(f"{project_name}/content")
    if not os.path.exists(f"{project_name}/content/items"):
        os.mkdir(f"{project_name}/content/items")
    if not os.path.exists(f"{project_name}/content/blocks"):
        os.mkdir(f"{project_name}/content/blocks")

    # MOD情報のHJSON書き込み
    if mod_info:
        mod_file_path = f"{project_name}/mod.hjson"
        with open(mod_file_path, 'w', encoding='utf-8') as f:
            hjson.dump(mod_info, f, ensure_ascii=False, indent=4)

    # アイテム情報のHJSON書き込み
    if item_info:
        item_file_path = f"{project_name}/content/items/{item_info['name']}.hjson"
        with open(item_file_path, 'w', encoding='utf-8') as f:
            hjson.dump(item_info, f, ensure_ascii=False, indent=4)

    # ブロック情報のHJSON書き込み
    if block_info:
        block_file_path = f"{project_name}/content/blocks/{block_info['name']}.hjson"
        with open(block_file_path, 'w', encoding='utf-8') as f:
            hjson.dump(block_info, f, ensure_ascii=False, indent=4)

# メイン関数
def main():
    project_name = input("プロジェクト名を入力してください: ")

    # プロジェクトのディレクトリとサブフォルダ作成
    if not os.path.exists(project_name):
        os.mkdir(project_name)
    if not os.path.exists(f"{project_name}/content"):
        os.mkdir(f"{project_name}/content")
    if not os.path.exists(f"{project_name}/content/items"):
        os.mkdir(f"{project_name}/content/items")
    if not os.path.exists(f"{project_name}/content/blocks"):
        os.mkdir(f"{project_name}/content/blocks")

    while True:
        print("\n1. MOD情報を入力")
        print("2. アイテム情報を入力")
        print("3. ブロック情報を入力")
        print("4. 終了")
        choice = input("選択してください: ")

        if choice == '1':
            mod_info = get_mod_info()
            write_to_hjson(project_name, mod_info, {}, {})
        elif choice == '2':
            item_info = get_item_info()
            write_to_hjson(project_name, {}, item_info, {})
        elif choice == '3':
            block_info = get_block_info()
            write_to_hjson(project_name, {}, {}, block_info)
        elif choice == '4':
            break
        else:
            print("無効な選択です。再度選んでください。")

if __name__ == "__main__":
    main()
```
